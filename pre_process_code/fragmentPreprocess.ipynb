```python
import numpy as np
import os
```

```python
directory = os.fsencode("../../samFile/")

with open("output.txt",'w') as newFile:
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.endswith(".sam"):
            fragLenDic = {}
            with open("../../samFile/"+filename,'r') as file1:

                #read 3 header lines
                line1=file1.readline().strip()
                line1=file1.readline().strip()
                line1=file1.readline().strip()
                line1=file1.readline().strip()
                while line1:
                    line1_sp = line1.split('\t')

                    cigarString = line1_sp[5]
                    fragLength = int(line1_sp[8])

                    bin = round(fragLength/400)
                    if bin < -100:
                        bin = -100
                    if bin > 100:
                        bin = 100

                    if bin not in fragLenDic:
                        fragLenDic[bin] = 1
                    else:
                        fragLenDic[bin] += 1

                    line1=file1.readline().strip()
            print(filename)
            newFile.write(filename+'\t')
            for i in range(-100,101):
                if i in fragLenDic:
                    newFile.write(str(fragLenDic[i]))
                else:
                    newFile.write(str(0))
                    
                if i != 100:
                    newFile.write('\t')
                else:
                    newFile.write('\n')
    
```
